```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].unique())
player_count
player_Analysis = pd.DataFrame({"Total Players": [player_count]})
player_Analysis
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items = len(purchase_data['Item Name'].unique())

average_price = purchase_data.iloc[:, 6:].mean(axis=1)
average_price = ('$ {:,.2f}'.format(average_price.mean()))

purchase_cnt = purchase_data.iloc[:, 6:].count(axis=1)
purchase_cnt = purchase_cnt.count()

total_purchase = purchase_data.iloc[:, 6:].sum(axis=1)
total_purchase = ('$ {:,.2f}'.format(total_purchase.sum()))

Unique_Items, average_price, purchase_cnt, total_purchase
Purchase_Analysis = pd.DataFrame({"Number of Unique Items ":[Unique_Items],
                                  "Average Price ":[average_price],
                                  "Number of Purchase ":[purchase_cnt],
                                  "Total Revenue ":[total_purchase]
                                })
Purchase_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_Male=purchase_data.loc[purchase_data['Gender'] == 'Male', ['SN']]
gender_Male = len(gender_Male['SN'].unique())
gender_Male_percentage = ('{0:.2f}%'.format(gender_Male/player_count*100))

gender_Female=purchase_data.loc[purchase_data['Gender'] == 'Female', ['SN']]
gender_Female = len(gender_Female['SN'].unique())
gender_Female_percentage = ('{0:.2f}%'.format(gender_Female/player_count*100))

gender_Other=purchase_data.loc[(purchase_data['Gender'] != 'Female') & (purchase_data['Gender'] != 'Male'), ['SN']]
gender_Other = len(gender_Other['SN'].unique())
gender_Other_percentage = ('{0:.2f}%'.format(gender_Other/player_count*100))

gender_Male, gender_Male_percentage, gender_Female, gender_Female_percentage, gender_Other, gender_Other_percentage
Gender_Demographic_Analysis = pd.DataFrame({" ":["Male", "Female", "Other/Non-Disclosed"],
                                  "Total Count ":[gender_Male,gender_Female,gender_Other],
                                  "Percentage of Players":[gender_Male_percentage,gender_Female_percentage,gender_Other_percentage]
                                }, index=['a', 'b','c'])

Gender_Demographic_Analysis = Gender_Demographic_Analysis.set_index(' ')
Gender_Demographic_Analysis
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
f_Price = purchase_data.loc[purchase_data['Gender'] == 'Female',['SN','Price']]
f_Count = f_Price.count(axis=1)
f_Count = f_Count.count()
f_Avg = f_Price.mean(axis = 1)
f_Avg = ('${:,.2f}'.format( f_Avg.mean()))
f_Total = f_Price.sum(axis =1)
f_Total = ('${:,.2f}'.format(f_Total.sum()))
f_GroupPurchase = f_Price.groupby("SN")
f_Avg_Purchase = f_GroupPurchase.sum().mean()
f_Avg_Purchase = f_Avg_Purchase

m_Price = purchase_data.loc[purchase_data['Gender'] == 'Male', ['SN','Price']]
m_Count = m_Price.count(axis=1)
m_Count = m_Count.count()
m_Avg = m_Price.mean(axis =1)
m_Avg = ('${:,.2f}'.format( m_Avg.mean()))
m_Total = m_Price.sum(axis =1)
m_Total = ('${:,.2f}'.format(m_Total.sum()))
m_GroupPurchase = m_Price.groupby("SN")
m_Avg_Purchase =m_GroupPurchase.sum().mean()
#m_Avg_Purchase = m_Avg_Purchase

o_Price = purchase_data.loc[(purchase_data['Gender'] != 'Female') & (purchase_data['Gender'] != 'Male'), ['SN','Price']]
o_Count = o_Price.count(axis =1)
o_Count = o_Count.count()
o_Avg = o_Price.mean(axis =1)
o_Avg = ('${:,.2f}'.format( o_Avg.mean()))
o_Total = o_Price.sum(axis=1)
o_Total = ('${:,.2f}'.format(o_Total.sum()))
o_GroupPurchase = o_Price.groupby("SN")
o_Avg_Purchase = o_GroupPurchase.sum().mean()
#o_Avg_Purchase = o_Avg_Purchase
f_Count, f_Avg, f_Total, f_Avg_Purchase, m_Count, m_Avg, m_Total, m_Avg_Purchase, o_Count, o_Avg,o_Total,o_Avg_Purchase

Gender_Purchase_Analysis = pd.DataFrame({" ":["Female", "Male", "Other/Non-Disclosed"],
                             "Purchase Count ":[f_Count,m_Count,o_Count],
                             "Average Purchase Price":[f_Avg, m_Avg, o_Avg],
                             "Total Purchase Value":[f_Total, m_Total, o_Total],
                             "Avg Total Purchase per Person":[f_Avg_Purchase, m_Avg_Purchase, o_Avg_Purchase]
                                }, index=['a', 'b','c'])

Gender_Purchase_Analysis = Gender_Purchase_Analysis.set_index(' ')
Gender_Purchase_Analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
Age_Category = pd.cut(purchase_data.Age, bins=[0,9,14,19,24,29,34,39,100], 
                                  labels = ['10<', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])
purchase_data.insert(2, "Age_Category", Age_Category)

TlCount_PerAgeCat = purchase_data['Age_Category'].value_counts()
Percent_PerAgeCat = TlCount_PerAgeCat/purchase_cnt*100
Percent_perAgeStr = Percent_PerAgeCat.round(2).astype(str) + '%'
Age_Demographic = pd.concat([TlCount_PerAgeCat, Percent_perAgeStr], axis='columns', sort =True)
column_names = ['Total Count','Percentage of Player']
Age_Demographic.columns=column_names
Age_Demographic
```

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
TlCount_PerAgeCat = purchase_data['Age_Category'].value_counts()
Avg_Purchase = purchase_data.groupby('Age_Category').mean()
TL_Purchase = purchase_data.groupby('Age_Category').sum()
Purchase_perPeson = purchase_data.groupby('Age_Category').sum().mean()
Age_Purchase = pd.concat([TlCount_PerAgeCat, Avg_Purchase['Price'], TL_Purchase['Price']], axis='columns',
                          sort =True) 
column_names = ['Purchase Count', 'Avg Purchase Price', 'Total Purchase Value']
Age_Purchase.columns = column_names
Age_Purchase
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Player = purchase_data.groupby(['SN'])
Player_Purchase = Player.count()
Player_Avg_Price = Player.mean()
Player_Total_Purchase = Player.sum()
Playser_Summary_df = pd.concat([Player_Purchase['Price'], Player_Avg_Price['Price'], Player_Total_Purchase['Price']], 
                               axis='columns', sort =True)
Playser_Summary_df.columns.values
column_names = ['Purchase Count', 'Avg Purchase Price', 'Total Purchase Value']
Playser_Summary_df.columns = column_names
Playser_Summary_df
Playser_Summary_df.sort_values(by='Total Purchase Value', ascending = False)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Item = purchase_data.groupby(['Item ID', 'Item Name'])
Item_Purchase = Item.count()
Item_Avg_Price = Item.mean()
Item_Total_Purchase = Item.sum()
Summary_df = pd.concat([Item_Purchase['Price'], Item_Avg_Price['Price'], Item_Total_Purchase['Price']], axis='columns',
                          sort =True)
Summary_df.columns.values
column_names = ['Purchase Count', 'Avg Purchase Price', 'Total Purchase Value']
Summary_df.columns = column_names
Summary_df
```
